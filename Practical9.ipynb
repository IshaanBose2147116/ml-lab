```python
import tensorflow as tf
```

```python
# implementing AND gate
x = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [[0], [0], [0], [1]]
```

```python
and_ann = tf.keras.models.Sequential()
```

```python
and_ann.add(tf.keras.layers.Dense(32, 'relu'))
```

```python
and_ann.add(tf.keras.layers.Dense(1, 'sigmoid'))
```

```python
and_ann.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# for AND
and_ann.fit(x, y, batch_size=32, epochs=100)
```
