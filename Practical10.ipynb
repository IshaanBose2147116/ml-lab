```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("./dataset/phishing_website_data.csv")
df
```

```python
new_df = df.drop(['id'], axis=1)
new_df
```

```python
X = new_df[["SSLfinal_State", "URL_of_Anchor"]]
Y = new_df.Result
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=1, test_size=0.3, stratify=Y)
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(6, 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(6, 'relu'))
```

```python
ann.add(tf.keras.layers.Dense(1, 'sigmoid'))
```

```python
ann.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(x_train, y_train, batch_size=32, epochs=100)
```

```python
ann.save("phishing_ann.h5")
```

```python
fraud_df = pd.read_csv("./dataset/categorical_data_on_fraud.csv")
fraud_df.age = pd.Series(LabelEncoder().fit_transform(fraud_df.age))
fraud_df.gender = pd.Series(LabelEncoder().fit_transform(fraud_df.gender))
```

```python
x2 = fraud_df[['age', 'gender', 'category', 'amount']]
y2 = fraud_df[['fraud']]

x2_res, y2_res = SMOTEENN(random_state=1).fit_resample(x2, y2)
```

```python
x2_train, x2_test, y2_train, y2_test = train_test_split(x2_res, y2_res, test_size=0.75, stratify=y2_res)
```

```python
fraud_ann.add(tf.keras.layers.Dense(6, 'relu'))
```

```python
fraud_ann.add(tf.keras.layers.Dense(1, 'sigmoid'))
```

```python
fraud_ann.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])
```
